### Objective:
To explore the Titanic dataset and gain insights using Python libraries like Pandas, Matplotlib, and Seaborn.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
# Dataset info
df.info()
```

```python
# Null values
df.isnull().sum()
```

```python
# Survival count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```
