```python
import matplotlib.pyplot as plt

# Sample country population data (2022 estimates)
countries = ['India', 'China', 'USA', 'Indonesia', 'Pakistan']
population = [1408, 1440, 331, 273, 231]  # in millions

# Plotting the bar chart
plt.figure(figsize=(10,6))
plt.bar(countries, population, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.title('Population of Top 5 Countries (2022 Estimate)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
